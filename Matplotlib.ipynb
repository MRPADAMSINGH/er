# Creating Your First Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plotting a graph
plt.plot(x, y)

# Assigning title to the graph
plt.title('Line graph')

# Labeling Y axis
plt.ylabel('Y axis')

# Labeling X axis
plt.xlabel('X axis')

# Display the result
plt.show()
```

# Pie Charts

```python
import matplotlib.pyplot as plt

sizes = [30, 45, 25]
labels = ['Slice 1', 'Slice 2', 'Slice 3']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')  # Creating a pie chart
```

```python
explode = (0, 0, 0.1)  # Explode the second slice (Slice 2)
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%')
# Save the plot as a PNG file
plt.savefig('pie.png', format='png')
```

```python
colors = ['red', 'blue', 'green']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
```

# Adding Labels, Titles, and Customization

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, label='Data')
plt.title('Line graph')  # Title
plt.xlabel('X axis')  # X-axis label
plt.ylabel('Y axis')  # Y-axis label

plt.legend()  # Legend
```

```python
title_font = {'family': 'serif', 'color': 'blue', 'size': 14}
label_font = {'family': 'sans-serif', 'color': 'green', 'size': 12}

plt.title('Customized Title', fontdict=title_font)
plt.xlabel('Customized X Label', fontdict=label_font)
plt.plot(x, y, label='Data', linestyle='--')  # Dashed line style
```

# Subplots

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 1, 2, 1]

fig, axes = plt.subplots(1, 2)  # Create a 1x2 grid of subplots
axes[0].plot(x, y1)
axes[1].plot(x, y2)
```

```python
fig, axes = plt.subplots(2, 2)  # Create a 2x2 grid of subplots
plt.subplots_adjust(wspace=0.4, hspace=0.4)  # Adjust spacing
axes[0, 0].plot(x, y1)
axes[0, 1].plot(x, y2, color='g')
axes[1, 0].plot(x, y1)
axes[1, 1].plot(x, y2)
```
